## From Text to Date

```python
df_comp.date = pd.to_datetime(df_comp.date, dayfirst=True)
```

```python
df_comp.head()
```

```python
df_comp.date.describe()
```

## Setting the Index

```python
df_comp.set_index('date', inplace=True)
```

```python
df_comp.head()
```

## Setting the Frequency

```python
df_comp = df_comp.asfreq('b')
# b - business days
# d - days
# a - annualy
```

```python
df_comp.head()
```

## Handling Missing Values

```python
df_comp.isna().sum()
```

```python
#backfilling заполняет пустоты значениями ближайшего заполненного периода
#frontfilling заполняет пустоты ближайшим предыдущим значением
# заполнять средним плохо для временных рядов
```

```python
df_comp.spx = df_comp.spx.fillna(method='ffill')
```

```python
df_comp.ftse = df_comp.ftse.fillna(method = 'bfill')
```

```python
df_comp.dax = df_comp.dax.fillna(df_comp.dax.mean())
```

```python
df_comp.nikkei = df_comp.nikkei.fillna(method='ffill')
```
